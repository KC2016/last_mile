```python
import pandas as pd

df_map = pd.read_csv('data/grocery_cleaned_coordinates.csv')
```

```python
df_map.head()
```

```python
import folium

map_center = [df_map['drop_latitude'].mean(), df_map['drop_longitude'].mean()]
m = folium.Map(location=map_center, zoom_start=5)

for i, row in df_map.head(100).iterrows():
    folium.CircleMarker(
        location=[row['drop_latitude'], row['drop_longitude']],
        radius=2,
        color='blue',
        fill=True,
        fill_opacity=0.5
    ).add_to(m)

m.save('maps/sample_locations_map.html')
```
